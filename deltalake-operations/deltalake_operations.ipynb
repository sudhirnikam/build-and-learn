```python
from pyspark.sql import Row
from datetime import date
from decimal import Decimal
from pyspark.sql.types import DecimalType
from pyspark.sql.functions import col


new_data = [
    Row(order_id=2001, customer_id=601, order_date=date(2026,5,10),
        amount=Decimal("199.99"), status="COMPLETED", payment_method="CREDIT_CARD"),
    Row(order_id=2002, customer_id=602, order_date=date(2026,5,11),
        amount=Decimal("450.50"), status="PENDING",   payment_method="UPI"),
]

df = spark.createDataFrame(new_data) \
    .withColumn("amount", col("amount").cast(DecimalType(10, 2))) 
df.printSchema()
```

```python
df.write.format("delta").mode("append").saveAsTable("orders")
```

```python
df.write.format("delta") \
    .mode("append") \
    .option("mergeSchema", "true") \
    .saveAsTable("orders")
```
